```python
import tensorflow as tf
# def create_model(): 
# Step 1 - Shape
inputs=tf.keras.Input(shape = (image_size, image_size, 3))
#CNN
cnn1=tf.keras.layers.Convolution2D(filters = 128, kernel_size = (3, 3), activation = 'relu')(inputs)
# Adding a second convolutional layer
cnn2=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (3, 3), activation = 'relu')(cnn1)
# Step 2 - Pooling
block_1_output=tf.keras.layers.MaxPooling2D(pool_size = (2, 2))(cnn2)
# dropout=tf.keras.layers.Dropout(0.3)(block_1_output)


# Adding a second convolutional layer
cnn3=tf.keras.layers.Convolution2D(filters = 32, kernel_size = (3, 3), activation = 'relu',padding='same')(block_1_output)
max2=tf.keras.layers.MaxPooling2D(pool_size = (2, 2))(cnn3)
block_2_output = tf.keras.layers.add([max2, block_1_output])

# Adding a third convolutional layer
cnn4=tf.keras.layers.Convolution2D(filters = 16, kernel_size = (3, 3), activation = 'relu',padding='same')(block_2_output)
max3=tf.keras.layers.MaxPooling2D(pool_size = (2, 2))(cnn4)
#         model=tf.keras.layers.Dropout(0.3)(model)
# x = layers.GlobalAveragePooling2D()(x)
block_3_output = tf.keras.layers.add([max3, block_2_output])
# Step 3 - Flattening
model=tf.keras.layers.Flatten()(block_3_output)

# Step 4 - Full connection
model=tf.keras.layers.Dense(512, activation = 'relu',input_shape=(50176,))(model)
model=tf.keras.layers.Dropout(0.2)(model)
outputs=tf.keras.layers.Dense(1, activation = 'sigmoid',input_shape=(50176,))(model)
model = tf.keras.Model(inputs, outputs, name='vapacnn')
model.summary()
#         return model

```

```python
#create a basic model instance
# model = create_model()
# Compiling the CNN  loss = 'binary_crossentropy','categorical_crossentropy'
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
model.summary()
```

```python
import keras
from keras import layers
inputs = keras.Input(shape=(32, 32, 3), name='img')
x = layers.Conv2D(32, 3, activation='relu')(inputs)
x = layers.Conv2D(64, 3, activation='relu')(x)
block_1_output = layers.MaxPooling2D(2)(x)

x = layers.Conv2D(64, 3, activation='relu', padding='same')(block_1_output)
x = layers.Conv2D(64, 3, activation='relu', padding='same')(x)
block_2_output = layers.add([x, block_1_output])

x = layers.Conv2D(64, 3, activation='relu', padding='same')(block_2_output)
x = layers.Conv2D(64, 3, activation='relu', padding='same')(x)
block_3_output = layers.add([x, block_2_output])

x = layers.Conv2D(64, 3, activation='relu')(block_3_output)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(256, activation='relu')(x)
x = layers.Dropout(0.5)(x)
outputs = layers.Dense(10, activation='softmax')(x)

model = keras.Model(inputs, outputs, name='toy_resnet')
model.summary()
```

```python
import tensorflow as tf
# def create_model(): 
# Step 1 - Shape
inputs=tf.keras.Input(shape = (128, 128, 3))
#CNN
cnn1=tf.keras.layers.Convolution2D(filters = 128, kernel_size = (7, 7),strides=2, activation = 'relu')(inputs)
block_1_output=tf.keras.layers.MaxPooling2D(
    pool_size=(3, 3),
    strides=2,
    padding='same',
    data_format=None,

)(cnn1)
block_1_output=tf.keras.layers.BatchNormalization(axis=-1, momentum=0.99, 
                                                  epsilon=0.001, center=True,
                                                  scale=True, beta_initializer='zeros',
                                                  gamma_initializer='ones',
                                                  moving_mean_initializer='zeros',
                                                  moving_variance_initializer='ones',
                                                  beta_regularizer=None, gamma_regularizer=None,
                                                  beta_constraint=None, gamma_constraint=None)(block_1_output)


cnn2=tf.keras.layers.Convolution2D(filters = 128, kernel_size = (1, 1), activation = 'relu',padding='valid')(block_1_output)
lrn=tf.keras.layers.BatchNormalization(axis=-1, momentum=0.99, 
                                                  epsilon=0.001, center=True,
                                                  scale=True, beta_initializer='zeros',
                                                  gamma_initializer='ones',
                                                  moving_mean_initializer='zeros',
                                                  moving_variance_initializer='ones',
                                                  beta_regularizer=None, gamma_regularizer=None,
                                                  beta_constraint=None, gamma_constraint=None)(cnn2)
block_2_output=tf.keras.layers.MaxPooling2D(
    pool_size=(3, 3),
    strides=2,
    padding='same',
    data_format=None,

)(lrn)


############splitting

cnnsp1=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (3, 3),strides=1, activation = 'relu')(block_2_output)

cnnsp2=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (1, 1),strides=1, activation = 'relu')(block_2_output)
cnnsp2a=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (3, 3),strides=1, activation = 'relu')(cnnsp2)

cnnsp3=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (1, 1),strides=1, activation = 'relu')(block_2_output)
cnnsp3a=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (3, 3),strides=1, activation = 'relu')(cnnsp3)


maxps=tf.keras.layers.MaxPooling2D(
    pool_size=(3, 3),
    strides=1,
    padding='same',
    data_format=None,

)(block_2_output)
cnnmaxps=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (3, 3),strides=1, activation = 'relu')(maxps)

# lis=([cnnsp1,cnnsp2a,cnnsp3a,cnnmaxps])

concate = tf.keras.layers.concatenate(inputs=([cnnsp1,cnnsp2a,cnnsp3a,cnnmaxps]))



############splitting2

cnn2sp1=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (3, 3),strides=1, activation = 'relu')(concate)

cnn2sp2=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (1, 1),strides=1, activation = 'relu')(concate)
cnn2sp2a=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (3, 3),strides=1, activation = 'relu')(cnn2sp2)

cnn2sp3=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (1, 1),strides=1, activation = 'relu')(concate)
cnn2sp3a=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (3, 3),strides=1, activation = 'relu')(cnn2sp3)


max2ps=tf.keras.layers.MaxPooling2D(
    pool_size=(3, 3),
    strides=1,
    padding='same',
    data_format=None,

)(concate)
cnn2maxps=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (3, 3),strides=1, activation = 'relu')(max2ps)



concate2 = tf.keras.layers.concatenate(inputs=([cnn2sp1,cnn2sp2a,cnn2sp3a,cnn2maxps]))

###concated to max
maxcm=tf.keras.layers.MaxPooling2D(
    pool_size=(3, 3),
    strides=2,
    padding='same',
    data_format=None,

)(concate2)


############splitting3 ###############################################3

cnn3sp1=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (3, 3),strides=1, activation = 'relu')(maxcm)

cnn3sp2=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (1, 1),strides=1, activation = 'relu')(maxcm)
cnn3sp2a=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (3, 3),strides=1, activation = 'relu')(cnn3sp2)

cnn3sp3=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (1, 1),strides=1, activation = 'relu')(maxcm)
cnn3sp3a=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (3, 3),strides=1, activation = 'relu')(cnn3sp3)


max3ps=tf.keras.layers.MaxPooling2D(
    pool_size=(3, 3),
    strides=1,
    padding='same',
    data_format=None,

)(maxcm)
cnn3maxps=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (3, 3),strides=1, activation = 'relu')(max3ps)



concate3 = tf.keras.layers.concatenate(inputs=([cnn3sp1,cnn3sp2a,cnn3sp3a,cnn3maxps]))

###concate 4  + average pool ########################

############splitting3 ###############################################3

cnn4sp1=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (3, 3),strides=1, activation = 'relu')(concate3)

cnn4sp2=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (1, 1),strides=1, activation = 'relu')(concate3)
cnn4sp2a=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (3, 3),strides=1, activation = 'relu')(cnn4sp2)

cnn4sp3=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (1, 1),strides=1, activation = 'relu')(concate3)
cnn4sp3a=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (3, 3),strides=1, activation = 'relu')(cnn4sp3)


max4ps=tf.keras.layers.MaxPooling2D(
    pool_size=(3, 3),
    strides=1,
    padding='same',
    data_format=None,

)(concate3)
cnn4maxps=tf.keras.layers.Convolution2D(filters = 64, kernel_size = (3, 3),strides=1, activation = 'relu')(max4ps)



concate4 = tf.keras.layers.concatenate(inputs=([cnn4sp1,cnn4sp2a,cnn4sp3a,cnn4maxps]))
# concate4 = tf.keras.layers.Merge([cnn4sp1,cnn4sp2a,cnn4sp3a,cnn4maxps], mode='concat', concat_axis=channel_axis)

# avg = tf.keras.layers.AveragePooling2D(pool_size=5, strides=None, padding='same', data_format='channels_last')(concate3)


########################################33 concate4 + avg ################

# add = tf.keras.layers.add(inputs=([concate4,avg]))

#######################################################################################################################
########avg + concate ##########



# # Step 3 - Flattening
# model=tf.keras.layers.Flatten()(block_3_output)

# # Step 4 - Full connection
# model=tf.keras.layers.Dense(512, activation = 'relu',input_shape=(50176,))(model)
# model=tf.keras.layers.Dropout(0.2)(model)
# outputs=tf.keras.layers.Dense(1, activation = 'sigmoid',input_shape=(50176,))(model)
model = tf.keras.Model(inputs, concate4, name='incnn')
model.summary()
tf.keras.utils.plot_model(model, 'own_inception_model.png',show_shapes=True)
#         return model

```
